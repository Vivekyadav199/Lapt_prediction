## Now we creat 4 columns for the Memory
- ssd
- hdd
- flash storage
- hybrid

```python
lapt.head()
```

```python
lapt['Memory'] = lapt['Memory'].astype('str').replace('\.0', '', regex=True)
lapt["Memory"] = lapt["Memory"].str.replace('GB', '')
lapt["Memory"] = lapt["Memory"].str.replace('TB', '000')
new = lapt["Memory"].str.split("+", n = 1, expand = True)
#new = df["Memory"].str.split("+", n = 1, expand = True)
## (.str.split("+", n = 1, expand = True): This part of the code applies the split method to the selected column, splitting each value by the "+" character. The n = 1 parameter specifies that it should only split at the first occurrence of "+". The expand = True parameter tells pandas to create a new DataFrame with split values.)
## regex = True (parameter indicates that the pattern should be treated as a regular expression.
lapt["first"]= new[0]
lapt["first"]=lapt["first"].str.strip()

lapt["second"]= new[1]

lapt["Layer1HDD"] = lapt["first"].apply(lambda x: 1 if "HDD" in x else 0)
lapt["Layer1SSD"] = lapt["first"].apply(lambda x: 1 if "SSD" in x else 0)
lapt["Layer1Hybrid"] = lapt["first"].apply(lambda x: 1 if "Hybrid" in x else 0)
lapt["Layer1Flash_Storage"] = lapt["first"].apply(lambda x: 1 if "Flash Storage" in x else 0)

lapt['first'] = lapt['first'].str.replace(r'\D', '')

lapt["second"].fillna("0", inplace = True)

lapt["Layer2HDD"] = lapt["second"].apply(lambda x: 1 if "HDD" in x else 0)
lapt["Layer2SSD"] = lapt["second"].apply(lambda x: 1 if "SSD" in x else 0)
lapt["Layer2Hybrid"] = lapt["second"].apply(lambda x: 1 if "Hybrid" in x else 0)
lapt["Layer2Flash_Storage"] = lapt["second"].apply(lambda x: 1 if "Flash Storage" in x else 0)

lapt['second'] = lapt['second'].str.replace(r'\D', '')

lapt["first"] = lapt["first"].astype(int)
lapt["second"] = lapt["second"].astype(int)

lapt["HDD"]=(lapt["first"]*lapt["Layer1HDD"]+lapt["second"]*lapt["Layer2HDD"])
lapt["SSD"]=(lapt["first"]*lapt["Layer1SSD"]+lapt["second"]*lapt["Layer2SSD"])
lapt["Hybrid"]=(lapt["first"]*lapt["Layer1Hybrid"]+lapt["second"]*lapt["Layer2Hybrid"])
lapt["Flash_Storage"]=(lapt["first"]*lapt["Layer1Flash_Storage"]+lapt["second"]*lapt["Layer2Flash_Storage"])

lapt.drop(columns=['first', 'second', 'Layer1HDD', 'Layer1SSD', 'Layer1Hybrid',
       'Layer1Flash_Storage', 'Layer2HDD', 'Layer2SSD', 'Layer2Hybrid',
       'Layer2Flash_Storage'],inplace=True)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=0)
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.metrics import r2_score,mean_absolute_error
```

```python
from sklearn.linear_model import LinearRegression,Ridge,Lasso
from sklearn.ensemble import RandomForestRegressor
from sklearn.preprocessing import OneHotEncoder
```

```python
step1 = ColumnTransformer(transformers =[
    ('col_tnf',OneHotEncoder(sparse = False, drop = 'first'),[0,1,7,10,11])
],remainder = 'passthrough' )
```

```python
step2 = LinearRegression()

pipe = Pipeline([
    ('step1',step1),
    ('step2',step2)
])
pipe.fit(x_train,y_train)
y_pred = pipe.predict(x_test)
print('R2 score',r2_score(y_test,y_pred))
print('MAE', mean_absolute_error(y_test,y_pred))
```
